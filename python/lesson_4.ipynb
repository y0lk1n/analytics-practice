```python
# Import file from CSV
path_to_file = '/mnt/HC_Volume_18315164/home-jupyter/jupyter-i-sedelkin/practice/data/3_taxi_peru.csv'
taxi = pd.read_csv(path_to_file, sep=';')
taxi
```

```python
# Convert values in columns to datetime one-by-one
taxi['start_at'] = pd.to_datetime(taxi['start_at'])
taxi['end_at'] = pd.to_datetime(taxi['end_at'])
taxi['arrived_at'] = pd.to_datetime(taxi['arrived_at'])

# Filter df by values in one of the columns
start_types_of_interest = ['asap', 'reserved']
taxi = taxi.loc[taxi.start_type.isin(start_types_of_interest)]
```

```python
# Get wait time from order to arrival

taxi['wait_time'] = (taxi.arrived_at - taxi.start_at).astype('timedelta64[m]')
```

```python
taxi.head()
```

```python
# Get drivers with the most times that they have been late

late_drivers = taxi \
    .query("wait_time > 0 & start_type == 'reserved'") \
    .groupby('driver_id', as_index=False) \
    .size() \
    .sort_values(by='size', ascending=False)
late_drivers
```

```python
# Extract arguments from date

taxi['month'] = taxi.start_at.dt.month
taxi['weekday'] = taxi.start_at.dt.strftime('%A')
```

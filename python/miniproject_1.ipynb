```python
# Function to batch-rename column names

def replace_spaces_capitals(obj):
    new_obj = obj \
        .replace(' ', '_') \
        .lower()
    return new_obj
```

```python
# Apply function

bookings = bookings.rename(columns=replace_spaces_capitals)
bookings.head()
```

```python
bookings.dtypes
```

```python
# Check number of successful entries

bookings \
    .query("is_canceled == 0").shape
```

```python
# Calculate top 5 countries by number of successful (not cancelled) reservations

bookings_successful_by_country = bookings \
    .query("is_canceled == 0") \
    .groupby('country', as_index=False) \
    .size() \
    .sort_values(by='size', ascending=False)

bookings_successful_by_country.head()
```

```python
# Compare to initial number of successful entries. I saw the discrepancy, but decided to ignore it for now

bookings_successful_by_country['size'].sum()
```

```python
# Calculate average number of nights booked in different hotel types

bookings \
    .groupby('hotel', as_index=False) \
    .agg({'stays_total_nights': 'mean'}) \
    .round(2)
```

```python
# Calculate number of cases where booked room type is different from assigned room type

bookings.query('reserved_room_type != assigned_room_type').shape
```

```python
# Find the most booked months for different years

year = 2017
bookings \
    .query('arrival_date_year == {}'.format(year)) \
    .groupby(['arrival_date_year', 'arrival_date_month'], as_index=False) \
    .size() \
    .sort_values(by='size', ascending=False)
```

```python
# Find the month with most cancellation for different years

year = 2015

bookings \
    .query("hotel == 'City Hotel' & arrival_date_year == {}".format(year)) \
    .groupby(['arrival_date_year', 'arrival_date_month'], as_index=False) \
    ['is_canceled'].sum() \
    .sort_values('is_canceled', ascending=False)
```

```python
# Find average number of visitors of different types in all bookings

bookings[['adults', 'children', 'babies']].mean()
```

```python
# Add column for total number of kids, find hotel type with mhigher average number of kids in booking

bookings['total_kids'] = bookings.children + bookings.babies
bookings.groupby('hotel', as_index=False).agg({'total_kids': 'mean'}).round(2)
```

```python
# Calculating churn by boolean value

bookings['has_kids'] = bookings.total_kids != 0

bookings_canceled_kids = bookings.query('is_canceled == 1').groupby(['has_kids', 'is_canceled'], as_index=False).agg('is_canceled').size()

bookings_kids = bookings.groupby(['has_kids'], as_index=False).size()
bookings_kids['canceled'] = bookings_canceled_kids['size']
bookings_kids['churn'] = (bookings_kids['canceled'] / bookings_kids['size'] * 100).round(2)
bookings_kids.head()
```

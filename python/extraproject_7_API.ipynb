```python
import requests
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
search_item = 'sunflower'

search_request_url = 'https://collectionapi.metmuseum.org/public/collection/v1/search?q={}'
r = requests.get(search_request_url.format(search_item))
res = r.json()
```

```python
sunflower_ids = [i for i in res['objectIDs']]
len(sunflower_ids)
```

```python
sunflower_ids
```

```python
reference_ids = [2032, 16822570, 437329, 436524, 20149]

for i in reference_ids:
    if i in(sunflower_ids):
        print(i)
    else:
        print('not a sunflower')
```

```python
target_id = 437980

object_request_url = 'https://collectionapi.metmuseum.org/public/collection/v1/objects/{}'
r_test = requests.get(object_request_url.format(target_id))

obj_test = r_test.json()
obj_test
```

```python
     objectID – id объекта
    title – название
    artistDisplayName – автор
    department – департамент
    objectBeginDate – дата (начало)
    objectEndDate – дата (конец)
    period – название периода
    objectName – название/категория объекта
    culture – культура
```

```python
sunflower_objects = pd.DataFrame(columns=['objectID', 'title', 'artistDisplayName', 'department', 'objectBeginDate', 'objectEndDate', 'period', 'objectName', 'culture'])
sunflower_objects
```

```python
index = 0

for i in sunflower_ids:
    
    target_id = i
    
    r_test = requests.get(object_request_url.format(target_id))
    obj_test = r_test.json()
    
    for n in sunflower_objects.columns:
        try:
            sunflower_objects.at[index, n] = obj_test[n]
        except:
            print('Could not add ' + str(i))
    
    print('added ' + str(i))
    index += 1
```

```python
sunflower_objects.reset_index(inplace=True)
```

```python
complex_search_request = 'https://collectionapi.metmuseum.org/public/collection/v1/search?departmentId={}&q={}'
department_id = 6
search_phrase = 'cat'

r = requests.get(complex_search_request.format(department_id, search_phrase))
cats = r.json()
```

```python
cat_ids = [i for i in cats['objectIDs']]
len(cat_ids)
```

```python
cat_objects = pd.DataFrame(columns=['objectID', 'title', 'artistDisplayName', 'department', 'objectBeginDate', 'objectEndDate', 'objectName', 'culture', 'period'])

index = 0

for i in cat_ids[:1000]:
    
    target_id = i
    
    r_test = requests.get(object_request_url.format(target_id))
    obj_test = r_test.json()
    
    for n in cat_objects.columns:
        try:
            cat_objects.at[index, n] = obj_test[n]
        except:
            print('Could not add ' + str(i))
    
    print('added ' + str(i))
    index += 1
```
